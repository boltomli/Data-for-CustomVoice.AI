Process text from Japanese digital library [青空文庫](https://www.aozora.gr.jp) to use with the audio from [青空朗読](http://aozoraroudoku.jp).

Clone the official repo `git clone https://github.com/aozorabunko/aozorabunko.git --depth=1 ../../aozorabunko` which holds all the text files such as `cards/000005/files/53194_45356.html`. We can traverse the files and get the texts with metadata for further use. Here we just show how to process one file.

```python
from bs4 import BeautifulSoup

doc = open('../../aozorabunko/cards/000005/files/53194_45356.html', 'rb')
content = doc.read()
soup = BeautifulSoup(content, 'lxml')

soup.find_all('div', class_='metadata')
```

Get the metadata so we can link text with the audio by title and author.

```python
soup.find_all('div', class_='main_text')
```

Notice that there are pronunciation labels for some of the kanji words. For example in `時計が<ruby><rb>零時半</rb><rp>（</rp><rt>れいじはん</rt><rp>）</rp></ruby>を打った。`, `零時半` should be read as `れいじはん`. So we may move forward to process the text further to have two versions, one with the original text and the other one partially transcribed to kana (to feed to NLP modules).
